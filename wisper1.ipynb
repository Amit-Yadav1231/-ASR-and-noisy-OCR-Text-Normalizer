```python
import whisper
model = whisper.load_model("base")
```

```python
model.device
```

```python
from IPython.display import Audio
Audio(filename="/content/narasimha_dialogue.mp3")
```

```python
audio = whisper.load_audio("/content/narasimha_dialogue.mp3")
audio = whisper.pad_or_trim(audio)
mel= whisper.log_mel_spectrogram(audio).to(model.device)

probs = model.detect_language(mel)
# The language probabilities are in the second element of the 'probs' tuple
print(f"Detected language: {max(probs[1], key=probs[1].get)}")
# now here i am ddcoding the Audio
lang=whisper.DecodingOptions()
result = whisper.decode(model, mel, lang)

print(result.text)
```

```python
import whisper
import gradio as gr

model = whisper.load_model("base")

def transcribe(audio_path):
    result = model.transcribe(audio_path)
    return result["text"]

gr.Interface(
    fn=transcribe,
    inputs=gr.Audio(type="filepath"),
    outputs=gr.Textbox(),
    title="OPENAI ASR",
).launch()
```
